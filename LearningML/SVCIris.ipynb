```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
n = 20
X = np.arange(n)
y = 4*X + 3*X**2-100
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m = LinearRegression()
m.fit(X[:,np.newaxis],y)
y_pred = m.predict(X[:,np.newaxis])
plt.scatter(X,y)
plt.plot(X,y_pred, color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly_model.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)
plt.scatter(X,y)
plt.plot(X,y,color='red')
```

```python
# inspect the coefficents
m.coef_
```

```python
X = sns.load_dataset('iris')
```

```python
X.head()
```

```python
y = X.species
X = X.drop(columns={'species'})
# Split data set in training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Support Vector Classification
model = SVC(gamma='auto') # auto = 1/n_features
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)
print(f"Predicted Iris species with an accuracy score of {round(accuracy*100, 2)}%")
```
