```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/adult/adult.data')
df.columns = ['age',
              'workclass',
              'fnlwgt',
              'education',
              'education-num',
              'martial-status',
              'occupation',
              'relationship',
              'race',
              'sex',
              'capital-gain',
              'capital-loss',
              'hours-per-week',
              'native-country',
              'salary']
```

```python
df.head(5)
```

```python
df.describe()
```

```python
# onehot encode categorical columns
df_encoded = pd.get_dummies(df, columns=['workclass', 'education', 'martial-status', 'occupation', 'relationship', 'race', 'sex', 'native-country'])
```

```python
df_encoded.head(5)
```

```python
df_encoded.shape
```

```python
y = df_encoded['salary']
```

```python
y
```

```python
salary_values = np.unique(y)
# get mask where salaray correspond to a value
idx = df_encoded.loc[:,'salary'] == salary_values[0]
```

```python
salary_values
```

```python
# replace salary values to 0 and 1. 0 indicates <=50K 1 >50K
df_encoded.loc[idx,:] = 0
df_encoded.loc[~idx,:] = 1
```

```python
# check
np.unique(df_encoded['salary'])
```

```python
df_encoded.to_csv('../datasets/adult/adult_onehot.csv')
```
